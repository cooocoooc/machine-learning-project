## Python kernel

```python
import json
from jupyter_client import find_connection_file

def get_current_kernel(): 
    cf = find_connection_file()
    with open(cf) as f:
        info = json.load(f)
    print("Kernel name:", info.get("kernel_name"))
```

```python
get_current_kernel()
```
